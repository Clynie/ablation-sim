```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sns.set_context('talk')
sns.set_style('darkgrid')
```

```python
data = np.load('scores.npy')
```

```python
data.shape
```

```python
c = sns.color_palette('Set2', 4)
sns.palplot(c)
plt.figure()
for i in xrange(4):
    plt.figure()
    diff = data[:,i+4,:] - data[:,i,:]
    avg = np.mean(diff, axis=0)
    dev = 2*np.std(diff, axis=0)
    plt.fill_between(np.arange(len(avg)), avg-dev, avg+dev, color=c[i], alpha=0.3)
    plt.plot(diff.T, 'k', alpha=0.05)
    plt.plot(diff[0], 'k', alpha=0.8)
    plt.plot(avg, color=c[i])
```
